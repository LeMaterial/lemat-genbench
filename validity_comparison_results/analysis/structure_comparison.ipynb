```python
import glob

import matplotlib.pyplot as plt
import pandas as pd
from pymatviz.ptable import ptable_heatmap_plotly
```

```python
with open("composition_analysis.txt", "r") as f:
    lines = [line.strip() for line in f]
```

```python
import re
```

```python
compositions = []
reduced_comp = []
count = 0
count_el = 0
el = "Si"
for line in lines:
    if count >0:
        if len(line) == 0:
            break
        if "-" not in line: 
            # print(line)
            compositions.append(line)
            if el in line:
                count_el += 1
    if "Ours Valid" in line: 
        count = 1


print(len(compositions))
print(count_el)
for comp in compositions:
    reduced = re.sub(r"\d+", "", comp)  # 'CsBiTe'
    reduced_comp.append(reduced)
```
