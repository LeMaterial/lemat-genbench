```python
import glob

import matplotlib.pyplot as plt
import pandas as pd
from pymatviz import ptable_heatmap_plotly
```

```python
with open("composition_analysis.txt", "r") as f:
    lines = [line.strip() for line in f]
```

```python
compositions = []
count = 0
for line in lines:
    if count >0:
        if len(line) == 0:
            break
        if "-" not in line: 
            print(line)
            compositions.append(line)
    if "Ours Valid" in line: 
        count = 1


print(len(compositions))
```

```python
ptable_heatmap_plotly(compositions)
```
